## Motivación

La motivación detrás del método aqui empleado fue combinar las dos predicciones de dos funciones de similitud léxica. Se usaron como base Jaro, Cosine Similarity y Word2vec. La predicción fue hecha con la siguiente formula:  
$prediccion = c_{1}*prediccion_{1} + c_{2}*prediccion_{2}$ donde $c_1 = [0, 0.25, 0.5, 0.75, 1]$ y $c_2 = 1-c_1$

En base a la mejor combinación de parámetros se reporta el coeficiente pearson sobre las predicciones y el Gold Standard. En algunos casos se puede destacar la mejora generada por darle mas importancia a un método como word2vec que genera similitud semántica.

```python
def evaluate_textual(fun1, fun2):
    print "DATASET\t#pares\tPearson r"
    suma_Pearson_r=0.0
    suma_numero_de_pares=0
    suma_Pearson_r_year={"2012":0.0,"2013":0.0,"2014":0.0,"2015":0.0,"2016":0.0}
    suma_numero_de_pares_year={"2012":0,"2013":0,"2014":0,"2015":0,"2016":0}

    bestparameter_dict = {}
    list_pearson = {}
    a = np.linspace(0,1,5)
    b = 1 - np.linspace(0,1,5)
    cparameters = zip(a, b)
    for dataset in datasets:
        predicciones1=[]
        predicciones2=[]
        d,gs=leer_dataset_textsim(dataset)
        for texto1,texto2 in d:
            # EJEMPLOS
            #prediccion=STS_monge_elkan(texto1,texto2,lexsim.lex_sim_jaccard)
            prediccion1=STS_monge_elkan(texto1,texto2,fun1)
            #prediccion3=STS_monge_elkan(texto1,texto2,lexsim.lex_sim_cosine)
            #prediccion=STS_monge_elkan(texto1,texto2,lexsim.lex_sim_path_edit_distance) # ojo, demora muchísimo
            prediccion2=STS_monge_elkan(texto1,texto2,fun2)
            predicciones1.append(prediccion1)
            predicciones2.append(prediccion2)
        bestp, pearson_r = evaluate_combination(cparameters, gs, predicciones1, predicciones2)
        prediccion = cparameters[bestp][0]*np.array(predicciones1) + cparameters[bestp][1]*np.array(predicciones2)
        predicciones.append(prediccion)
        Pearson_r=pearson_r[bestp]
        print dataset,"\t",len(d),"\t",round(Pearson_r,4)
        
        bestparameter_dict[dataset] = cparameters[bestp]
        list_pearson[dataset] = pearson_r
        
        suma_Pearson_r+=Pearson_r*len(d)
        suma_numero_de_pares+=len(d)
        #actualiza el promedio ponderado por año
        for year in suma_Pearson_r_year:
            if year in dataset:
                suma_Pearson_r_year[year]+=Pearson_r*len(d)
                suma_numero_de_pares_year[year]+=len(d)
         

    print "Promedio ponderado todos\t",suma_numero_de_pares,"\t",round(suma_Pearson_r/suma_numero_de_pares,4)

    mejor_en_SemEval={
        "2012":"0.6773 UKP",
        "2013":"0.6181 UMBC",
        "2014":"0.7610 DLS@CU",
        "2015":"0.8015 DLS@CU",
        "2016":"0.7781 Samsung Poland NLP",
        }
    for year in ["2012","2013","2014","2015","2016"]:
        print "Promedio ponderado",year,"\t",suma_numero_de_pares_year[year],"\t",round(suma_Pearson_r_year[year]/suma_numero_de_pares_year[year],4)    ,"\tMejor en SemEval:\t",mejor_en_SemEval[year]

    return bestparameter_dict, list_pearson
```

```python
def evaluate_combination(cparameters, gs, prediccion1, prediccion2):
    pearson_r = []
    for c1, c2 in cparameters:
        prediccion = c1*np.array(prediccion1)+ c2*np.array(prediccion2)
        pearson_r.append(pearsonr(gs,prediccion)[0])
    return np.argmax(pearson_r), pearson_r
```

```python
def plot_combinations(datasets, list_pearson):
    plt.figure(figsize=(8, 6))
    for dataset in datasets:
        scores = list_pearson[dataset]
        line1, = plt.plot(scores, label=dataset)
    plt.xlabel('C')
    plt.ylabel('Accuracy')
    plt.xticks(np.arange(len(cparameters)), cparameters, rotation=45)
    plt.title('Jaro - Sim Cosine')
    plt.legend(bbox_to_anchor=(1.6, 0.9), bbox_transform=plt.gcf().transFigure)
    plt.show()
```

In the following setup there is a combination between Jaro and word2vec

```python
bestparameter_dict, list_pearson = evaluate_textual(lexsim.lex_sim_Jaro, lexsim.lex_sim_word2vec)
plot_combinations(datasets, list_pearson)
```
